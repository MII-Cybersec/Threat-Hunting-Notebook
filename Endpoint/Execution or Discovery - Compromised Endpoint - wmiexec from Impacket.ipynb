# Execution or Discovery - Compromised Endpoint - wmiexec from Impacket

#### Author: Rifqi Ardia Ramadhan

WMI is a Windows administration feature that provides a uniform environment for local and remote access to Windows system components. It relies on the WMI service for local and remote access and the server message block (SMB) and Remote Procedure Call Service (RPCS) for remote access. `wmiexec.py` utilizes compromised endpoint such as DCOM interface, that can silently execute arbitrary commands with semi-shell spawning. 

## Use Case
### Objective
Identify if there is any suspicious WMI execution running in RPC protocol. This use case specifically try to detect from network traffic.

### Log Requirements
Network Traffic (DCOM/DCERPC)

### Characteristics
| No | Characteristic | Note | Why|
|----|----------------|------|---|
|  1 | `{8502C4DC-5FBB-11D2-AAC1-006008C78BC7}` UUID is shown on the TCP byte stream| | This UUID is identifier for `Win32_Process` for WMI Class|
|2   | `ROOT/cimv2` is shown in the TCP byte stream||`ROOT/cimv2` is a namespace used by WMI to interact with Windows Powershell|
|3   | `Win32_Process` is shown in the TCP byte stream | Optional | Similar to characteristics 1 |
