# Persistence - Persistent System Access Via Replacing onscreenkeyboard PE with cmd.exe
### Author: Bintang N M
The on screen keyboard PE is a part of windows's ease of access feature which accessible on the windows logon screen which when accessed will ran the osk.exe file. If an attacker replace the binary of osk.exe with cmd.exe, instead of spawning the on screen keyboard, it will open launch a shell as SYSTEM which has the equivalent ADMINISTRATOR privileges to run the net user commands.


## Use Case
### Objective
Identify anomalous on screen keyboard activity.


### Logs Requirements
Microsoft Windows Sysmon

### Filter for osk.exe and SYSTEM user

```python
evtx_file = "/logs/Persistence/persistence_accessibility_features_osk_sysmon1.evtx"

parser = evtx.PyEvtxParser(evtx_file)
parse_json = list(parser.records_json())

saved_event = []
# print(len(parse_json))

for pj in parse_json:
    if 'osk.exe' in pj['data'] and 'SYSTEM' in pj['data']:
        saved_event+=[pj]
        print(pj['data'])
```
