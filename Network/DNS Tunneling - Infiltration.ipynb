# DNS Tunneling - Infiltration

#### Author: Rifqi Ardia Ramadhan

DNS has the ability to query a domain with a TXT domain description embedded with it. But this ability can be abused for malware to infiltrate a machine, because usually DNS queries are ignored by network administrator. With this ability, malware can query a domain from a malicious DNS resolver, that will return malicious code embedded in TXT record.


Example C2 DNS query with TXT type response diagram from PaloAlto networks.

Picture above is a good illustration on how DNS infiltration works. This use case will try to detect DNS infiltration to prevent injection of malicious entity.


## Usecase
### Objective
Identify if there is any data infiltration that is tunneled from DNS service.

### Log Requirements
DNS log or traffic capture.

### Characteristics
| No | Characteristic | Note | Why|
|----|----------------|------|---|
|  1 | DNS TXT queries have unusually packet size and have almost fixed size.| | |
|2   | The content of TXT are unusual and related between several consecutive queries.|| Malicious entity is not that small to be included in the DNS TXT query|

### Sample
Dataset acquired from [PCAP-ATTACK](https://github.com/sbousseaden/PCAP-ATTACK)

The first useful filter on Wireshark is filtering only TXT queries with this filter:

```dns.resp.type == 16```

And then sort the packets by the biggest. If there are several large packets, check the content.
