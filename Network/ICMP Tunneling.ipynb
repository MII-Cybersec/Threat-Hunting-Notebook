# ICMP Tunneling

#### Author: Rifqi Ardia Ramadhan

ICMP has the ability to send arbitrary data over the network. For any type 0 (echo reply) or type 8 (echo message) ICMP packet allows a custom message to be sent for checking whether the request and response message match. This tunneling can be used to do infiltration against machines. Blank ICMP packets with no payload are usually 42 bytes. There are several ping method, some of them use several alphanumeric characters to be used for diagnostic purpose. Unusually large payload needs to be checked. This use case will try to detect arbitrary messages that is sent from ICMP.

## Use Case
### Objective
Identify if there is any suspicious ICMP packets that are carrying unusual payload.

### Log Requirements
ICMP log or network traffic capture.

### Characteristics
| No | Characteristic | Note | Why|
|----|----------------|------|---|
|  1 | ICMP payload have unusually large packet size and have almost fixed size over several packets at a time.| | |
|2   | The content of the payload are unusual and related between several consecutive queries.|| Malicious entity is not that small to be included in the ICMP payload|
